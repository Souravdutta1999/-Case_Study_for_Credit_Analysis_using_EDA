## Handling missing values

### Dropping missing values 
- We can drop the columns which have more than 45% null values 

```python
percentage = 45.0 # Like N %
min_count =  int(((100-percentage)/100)*df1.shape[0] + 1)
new_df = df1.dropna( axis=1, 
                thresh = min_count)
```

```python
new_df.shape # Now we can see after dropping columns which have more than 45% null values 49 columns removed 
```

### Imputing missing values
- Filling the missing data with a value - Imputation

```python
# Checking null values % in OCCUPATION_TYPE column 
df1.OCCUPATION_TYPE.isnull().sum()*100/len(df1)
```

```python
# In OCCUPATION_TYPE column almost 31% null values & we don't know what should be inplace of null values, neither we can drop the column nor we can replace it with mode values.
# If we drop we won't get proper insight & if we use mode than it would be wrong approach.
# We can replace all the null values with Unknown category 
new_df['OCCUPATION_TYPE'] = new_df.OCCUPATION_TYPE.fillna('Unknown')
```

- Another best way for imputing missing values is: 1) using median for numeric columns 2) using mode for categorical columns

```python
for i in new_df.columns:
    if (new_df[i].dtype == 'float64') or (new_df[i].dtype == 'int64'):
        (new_df[i].fillna(new_df[i].median(), inplace = True))
    else:
        (new_df[i].fillna(new_df[i].mode()[0],inplace = True))
       
```

## Top 10 Correlation for trgt_0 & trgt_1

```python
cont_cols = continious_cols.copy()
```

```python
# Select Upper triangle of correlation matrix
correlation_matrix = cont_cols.corr().abs()
upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(np.bool))
# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
```

```python
# Drop features 
cont_cols = cont_cols.drop(cont_cols[to_drop], axis=1)
```

```python
def get_redundant_pairs(new_df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = new_df.columns
    for i in range(0, new_df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(new_df, n=10):
    au_corr = new_df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(new_df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

print("Top 10 Absolute Correlations")
get_top_abs_correlations(cont_cols, 10).to_frame().style.background_gradient(cmap='coolwarm').set_precision(3)
```
